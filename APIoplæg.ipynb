```python
import requests

def fetch_users():
    url = "https://jsonplaceholder.typicode.com/users"
    response = requests.get(url)
    
    if response.status_code == 200:
        users = response.json()
        for user in users:
            print(f"Name: {user['name']}, Email: {user['email']}")
    else:
        print(f"Failed to retrieve data: {response.status_code}")

# Call the function to fetch and display user data
fetch_users()
```

```python
import requests

def explore_response_properties():
    # Make a GET request
    response = requests.get("https://api.github.com")

    # Check if the request was successful
    if response.status_code == 200:
        print("URL:", response.url)
        print("Status Code:", response.status_code)
        print("Headers:", response.headers)
        print("Encoding:", response.encoding)
        print("Elapsed Time:", response.elapsed)
        print("Cookies:", response.cookies)
        print("Text:", response.text[:100])  # print only the first 100 characters
    else:
        response.raise_for_status()

# Call the function
explore_response_properties()
```

```python
import requests

def fetch_data():
    # Make a GET request
    response = requests.get('https://jsonplaceholder.typicode.com/posts')

    # Check if the request was successful
    if response.status_code == 200:
        # Using list comprehension to create a summary of key attributes
        attributes = [
            ('Status Code', response.status_code),
            ('URL', response.url),
            ('Headers', response.headers['content-type']),
            ('Elapsed Time', response.elapsed),
            ('Encoding', response.encoding),
            ('First 100 characters of content', response.text[:100])
        ]
        
        # Print out each attribute name and its value
        for attr_name, attr_value in attributes:
            print(f"{attr_name}: {attr_value}")
    else:
        print("Failed to retrieve data:", response.status_code)

# Call the function
fetch_data()
```

```python
import requests

def dynamic_attribute_access():
    # Send a request
    response = requests.get('https://jsonplaceholder.typicode.com/posts')
    
    # Define the attributes you want to inspect
    attr_names = ['status_code', 'url', 'elapsed', 'encoding']
    
    # Use list comprehension to collect attribute values
    attributes = [(attr, getattr(response, attr)) for attr in attr_names if hasattr(response, attr)]

    # Print attributes
    for attr_name, attr_value in attributes:
        print(f"{attr_name}: {attr_value}")

# Execute the function
dynamic_attribute_access()
```

```python
import requests

def explore_all_attributes():
    # Send a request
    response = requests.get('https://jsonplaceholder.typicode.com/posts')

    # Use dir() to get a list of all attributes of the response object
    all_attributes = dir(response)

    # Filter out special methods/attributes (those that start with '__')
    attributes = {attr: getattr(response, attr) for attr in all_attributes if not attr.startswith('__')}

    # Print each attribute and its value, formatted for easier reading
    for attr_name, attr_value in attributes.items():
        print(f"{attr_name}: {attr_value}")

# Execute the function
explore_all_attributes()
```
